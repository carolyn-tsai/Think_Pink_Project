```python
import pandas as pd
data = pd.read_csv('machine_learning_data_test.csv', low_memory=False)
data.head()
```

```python
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("\'")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace('"',''))
    return (12*ft_) + in_
```

```python
height = data["Height (ft-in)"][1]
height
```

```python
parse_ht(height)
```

```python
data["Ht (inches)"] = data["Height (ft-in)"].apply(lambda x:parse_ht(x))
```

```python
data.head()
```

```python
def band_size(size):
    band = int(size[0:2])
    return band

def cup_size(size):
    cup = size[2:]
    return cup
```

```python
band_size("36AA")
```

```python
cup_size("36AA")
```

```python
data["Band Size"] = data["Bra Size"].apply(lambda x:band_size(x))
data["Cup Size"] = data["Bra Size"].apply(lambda x:cup_size(x))
```
