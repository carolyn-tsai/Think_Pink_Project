```python
# Import dependencies
from sklearn import tree
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load data
df = pd.read_csv(os.path.join("Machine_Learning_Data.csv"))
df.head()
```

```python
target = df["Bra Style"]
#target_names = ["negative", "positive"]
```

```python
data = df.drop("Bra Style", axis=1)
#data = data.drop("Height (ft-in)", axis=1)
#data = data.drop("Bra Size", axis=1)
feature_names = data.columns
data.head()
```

```python
# Create color cells 
colors = pd.get_dummies(data ["Bra Color"])
colors
# Insert color cells 
data = pd.concat([data, colors], axis =1)
data 
# Drop previous Bra Color column
data = data.drop("Bra Color", axis=1)
data
```

```python
# Create R/L/B/P cells 
rlbp = pd.get_dummies(data ["R/L/B/P"])
rlbp
# Insert R/L/B/P cells 
data = pd.concat([data, rlbp], axis =1)
data 
# Drop previous R/L/B/P Color column
data = data.drop("R/L/B/P", axis=1)
data
```

```python
# Create Brand cells 
brand = pd.get_dummies(data ["Brand"])
brand
# Insert Brand cells 
data = pd.concat([data, brand], axis =1)
data 
# Drop previous Brand column
data = data.drop("Brand", axis=1)
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# data = pd.concat([data, .......])
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
